```python
import numpy as np
import matplotlib.pyplot as plt
```

$$ f(x) = 0.2 + 25x - 200x^2 + 675x^3 - 900x^4 + 400x^5 $$

$$ I = \int_{0}^{0.8} f(x) dx$$

Actual value of $ I $ is $ 1.640533 $.


```python
def trapM(x, y):
    I = 0
    for i in range(1, len(x)):
        h = x[i] - x[i-1]
        I += h*(y(x[i]) + y(x[i-1]))/2
    return I
```

```python
# Example
def y1(x):
    return 0.2 + 25*x - 200*x**2 + 675*x**3 - 900*x**4 + 400*x**5

x = np.linspace(0, 0.8, 10)
y = y1(x)
```

```python
I = trapM(x, y1)
I
```

```python
# Simpson's 1/3 Rule
def simpson13(x, y):
    # All the x must be equidistant
    I = 0
    for i in range(1, len(x), 2):
        h = x[i] - x[i-1]
        I += h*(y(x[i-2]) + 4*y(x[i-1]) + y(x[i]))/3
    return I
```

```python
# Simpson's 3/8 Rule
def simpson38(x, y):
    # All the x must be equidistant
    I = 0
    for i in range(1, len(x), 3):
        h = x[i] - x[i-1]
        I += 3*h*(y(x[i-3]) + 3*y(x[i-2]) + 3*y(x[i-1]) + y(x[i]))/8
    return I
```

```python
x = np.linspace(0, 0.8, 20)
y = y1(x)

I = simpson13(x, y1)
```
